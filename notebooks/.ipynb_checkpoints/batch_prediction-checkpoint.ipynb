Plot one unique_id (clean series)

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load  CSV
df = pd.read_csv("/home/rbahmani/projects/energy_mlops/artifacts/20251128_084045/batch_prediction/predictions.csv", parse_dates=["ds"])
df[df["unique_id"] == "group_1"]
```

```python

# 2. Pick one group to visualize (for example group_1)
uid = "group_1"
df_id = df[df["unique_id"] == uid].copy()
df_id = df_id.sort_values("ds")

# 3. Plot
plt.figure(figsize=(12, 6))
plt.plot(df_id["ds"], df_id["y_pred"], label="Actual", linewidth=2)

plt.title(f"Predicted ({uid})")
plt.xlabel("Time")
plt.ylabel("Energy consumption")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
